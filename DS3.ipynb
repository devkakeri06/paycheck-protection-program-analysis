```python
print(mdf.shape)
mdf.keys()
```

```python
# The national average loan amount.

print("The national avg loan amount is $", format(mdf["InitialApprovalAmount"].mean(), ",.2f"))
```

```python
print(mdf.BorrowerState.unique())
```

```python
# The average loan amount for every state.

mdf.groupby('BorrowerState', as_index=False)['InitialApprovalAmount'].mean()
```

```python
# Average loan amount by city. 

mdf.groupby('BorrowerCity', as_index=False)['InitialApprovalAmount'].mean()
```

```python
# mean loan amount grouped according to various KPIs

mdf.groupby(['BusinessType', 'Race', 'Ethnicity', 'Gender']).agg({'InitialApprovalAmount': ['mean']}).head(50)
# grouped_mean.head(50)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(16, 10))
plt.xlim(0, 150)
# plt.ylim(0.0, 0.4)
plt.scatter( mdf["JobsReported"], mdf["InitialApprovalAmount"], s=1)
plt.show()
```

```python
# mdf.groupby(['BorrowerCity']).agg({'InitialApprovalAmount': ['sum']})
# mdf.head()
mdf['date'] = pd.to_datetime(mdf['DateApproved'],format='%m/%d/%Y')
# df['year'] = pd.DatetimeIndex(df['date']).year
mdf['month'] = pd.DatetimeIndex(mdf['date']).month
# print(mdf['month'])
mdf.groupby(['month']).agg({'InitialApprovalAmount': ['sum']})
```

```python
# print(mdf.head())
mdf.sort_values(by='date', ascending=True, inplace=True)
# print(mdf.head(50))
```

**CONCLUSION**

Average National Loan Amount:- $ 27,396.83

There is no perfect linear relation between jobs reported and loan amount but on an average the lesser the jobs that were reported higher was the loan amount. so there is somewhat of a negative correlation.

Corporations reported most jobs and Rollover as Business Start-up (ROBS) reported the least jobs.

Housing CO-OP had the highest of the average loan amount and Self-Employed Individuals had the lowest average loan amount.

On an average men took more loan than women.

Month of April reported the most amount of loans.
